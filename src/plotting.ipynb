```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams

# Load your dataframe here, replacing 'your_data.xlsx' with the path to your data file
df = pd.read_excel('filename_20240129_015327.xlsx')

# Set the font to Times New Roman
rcParams['font.family'] = 'serif'
rcParams['font.serif'] = 'Times New Roman'
rcParams['font.size'] = 12

def plot_simulation_results(df):
    # Set the plot style
    sns.set(style="whitegrid")

    # Create a color palette with seaborn
    palette = sns.color_palette("husl", df['Protocol'].nunique())

    # Plot Throughput
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=df, x='Node', y='Average Throughput', hue='Protocol', style='Faulty Nodes', markers=True, ci='sd', palette=palette)
    plt.title('Throughput vs. Number of Nodes')
    plt.xlabel('Number of Nodes')
    plt.ylabel('Throughput (messages/s)')
    plt.legend(title='Protocol and Faulty Nodes')
    plt.tight_layout()
    plt.show()

    # Plot Latency
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=df, x='Node', y='Average Latency', hue='Protocol', style='Faulty Nodes', markers=True, ci='sd', palette=palette)
    plt.title('Latency vs. Number of Nodes')
    plt.xlabel('Number of Nodes')
    plt.ylabel('Latency (s)')
    plt.legend(title='Protocol and Faulty Nodes')
    plt.tight_layout()
    plt.show()

    # Plot CP Messages
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=df, x='Node', y='Average CP Messages', hue='Protocol', style='Faulty Nodes', markers=True, ci='sd', palette=palette)
    plt.title('CP Messages vs. Number of Nodes')
    plt.xlabel('Number of Nodes')
    plt.ylabel('CP Messages')
    plt.legend(title='Protocol and Faulty Nodes')
    plt.tight_layout()
    plt.show()

# Call the plotting function
plot_simulation_results(df)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
import pandas as pd

# Replace 'filename.xlsx' with the path to your Excel file
df = pd.read_excel('filename_20240129_015327.xlsx')
df = df[df['Faulty Nodes'] == 0]

def plot_simulation_results(df):
    with plt.style.context(['science', 'ieee']):
        # Plot Throughput
        fig, ax = plt.subplots(figsize=(12, 6))
        for protocol in df['Protocol'].unique():
            for type_ in df['Type'].unique():
                faulty_nodes=0
                subset = df[(df['Protocol'] == protocol) & (df['Faulty Nodes'] == faulty_nodes) & (df['Type'] == type_)]
                ax.plot(subset['Node'], subset['Average Throughput'], label=f'{protocol}, {faulty_nodes}, {type_}')
        ax.legend(title='Protocol, Faulty Nodes and Type')
        ax.set(xlabel='Number of Nodes', ylabel='Throughput (The number of transactions/s)')
        fig.savefig('figures/throughput.jpg', dpi=300)
        plt.close()

        # Plot Latency
        fig, ax = plt.subplots(figsize=(12, 6))
        for protocol in df['Protocol'].unique():
            for type_ in df['Type'].unique():
                faulty_nodes=0
                subset = df[(df['Protocol'] == protocol) & (df['Faulty Nodes'] == faulty_nodes) & (df['Type'] == type_)]
                ax.plot(subset['Node'], subset['Average Latency'], label=f'{protocol}, {faulty_nodes}, {type_}')
        ax.legend(title='Protocol, Faulty Nodes and Type')
        ax.set(xlabel='Number of Nodes', ylabel='Latency (s)')
        fig.savefig('figures/latency.jpg', dpi=300)
        plt.close()

        # Plot CP Messages
        fig, ax = plt.subplots(figsize=(12, 6))
        for protocol in df['Protocol'].unique():
            for type_ in df['Type'].unique():
                faulty_nodes=0
                subset = df[(df['Protocol'] == protocol) & (df['Faulty Nodes'] == faulty_nodes) & (df['Type'] == type_)]
                ax.plot(subset['Node'], subset['Average CP Messages'], label=f'{protocol}, {faulty_nodes}, {type_}')
        ax.legend(title='Protocol, Faulty Nodes and Type')
        ax.set(xlabel='Number of Nodes', ylabel='CP Messages')
        fig.savefig('figures/cp_messages.jpg', dpi=300)
        plt.close()

# Assuming `df` is your DataFrame and it contains the necessary columns
plot_simulation_results(df)
```
